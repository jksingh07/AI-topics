# Graphs

```python
import networkx as nx
```

```python
G = nx.Graph()

e = [(1,2), (2,3), (4,3), (5,2)]

G.add_edges_from(e)
```

```python
pos = nx.spring_layout(G)
```

```python
# WEIGHTED GRAPH

nx.draw(G, pos, labels = {n:n for n in G.nodes()})
l = {(1,2):10, (2,3):15, (4,3):25, (5,2):40}

nx.draw_networkx_edge_labels(G, pos, edge_labels=l);
```

## PATH GRAPH

```python
# Path graph -  end nodes has degree of 1 and nodes in between has degree of 2

G =nx.path_graph(G)
nx.draw_networkx(G)
```

## Star Graph

```python
G = nx.star_graph(8)
nx.draw_networkx(G)
```

## Cycle Graph

```python
G = nx.cycle_graph(15)
nx.draw_networkx(G)
```

## Complete Graph

```python
G = nx.complete_graph(G)
nx.draw_networkx(G)
```

```python
G = nx.complete_graph(5)
nx.draw_networkx(G)
```

### Special Graphs

```python
# Special graph of complete graph
G = nx.tetrahedral_graph() # only 4 nodes and each has 3 edges
nx.draw_networkx(G)
```

```python
G = nx.octahedral_graph()
nx.draw_networkx(G)
```

```python
# 10 nodes, nested stars
G = nx.petersen_graph()
nx.draw_shell(G, with_labels=True, nlist=[[5,6,7,8,9], [0,1,2,3,4]])
```

## Grid graph

```python
# 2D Graph

G = nx.grid_2d_graph(3,5)
nx.draw_networkx(G)
```

```python
# 3D Graph

G = nx.grid_graph([2,2,2])
nx.draw_networkx(G)
```

## Erdos Renyi

```python
G = nx.erdos_renyi_graph(7,0.5) # 7 nodes and 0.5 is the probability of edge b/w them
nx.draw_networkx(G)
```

```python
G = nx.Graph()

G.add_edge(1,2,w=0.2)
G.add_edge(3,2,w=0.3)
G.add_edge(4,2,w=0.7)
G.add_edge(5,2,w=0.8)

pos = nx.spring_layout(G) # Gives the actual location of nodes

nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_labels(G, pos, font_size=15);

a = [(u,v) for (u,v,d) in G.edges(data=True) if d['w']<=0.5]
b = [(u,v) for (u,v,d) in G.edges(data=True) if d['w']>0.5]

nx.draw_networkx_edges(G, pos, edgelist=a, style='dashed', edge_color='g')

nx.draw_networkx_edges(G, pos, edgelist=b, style='solid', edge_color='r')
```
